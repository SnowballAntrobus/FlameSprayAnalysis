```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load the dataset

```python
input_file = "data.csv"
df = pd.read_csv(input_file, header = 0)
```

### Split into input (X) and output (y) variables

```python
dataset = df.values
X = dataset[:,1:7]
y = dataset[:,7]
```

### Check if data is skewed

```python
result_freq = df["Result"].value_counts().plot.bar(title="Results")
result_freq.set_xticklabels(["Stable", "Unstable"], rotation="horizontal")
```

### Plotting histograms

```python
success = dataset[dataset[:,7] == 1]
failure = dataset[dataset[:,7] == 0]
```

```python
for idx, val in enumerate(list(df)):
    if idx == 0 or idx == 7:
        continue
    fig, ax = plt.subplots()
    ax.hist(success[:,idx], bins=10, alpha=0.5, label='Stable')
    ax.hist(failure[:,idx], bins=10, alpha=0.5, label='Unstable')
    ax.legend(loc='upper right')
    ax.set_xlabel(val)
    ax.set_ylabel("#")
    plt.show()
```
