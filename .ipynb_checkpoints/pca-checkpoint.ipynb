```python
import pandas as pd
from sklearn import preprocessing, decomposition, neighbors, model_selection
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

## Load the dataset

```python
input_file = "data.csv"
df = pd.read_csv(input_file, header = 0)
```

```python
dataset = df.values
X = dataset[:,1:7]
y = dataset[:,7]
y = y.astype('int')
```

## Apply PCA

Normalize data for PCA

```python
sc = preprocessing.StandardScaler()
X = sc.fit_transform(X)
```

### 2D

```python
pca = decomposition.PCA(n_components=2, random_state=1)
X_pca = pca.fit_transform(X)
```

```python
success = np.array([0.0,0.0])
failure = np.array([0.0,0.0])
for i in range(53):
    if y[i] == 0:
        failure = np.append(failure, X_pca[i])
    else:
        success = np.append(success, X_pca[i])
failure = failure.reshape((33,2))
success = success.reshape((22,2))
```

Stable: Orange, Unstable: Blue

```python
fig, ax = plt.subplots()
ax.scatter(failure[1:, 0], failure[1:, 1], alpha=0.8, marker="x")
ax.scatter(success[1:, 0], success[1:, 1], alpha=0.5, marker="^")
ax.set_title("PCA")
```

### 3D

```python
pca = decomposition.PCA(n_components=3, random_state=1)
X_pca = pca.fit_transform(X)
```

```python
success = np.array([0.0,0.0,0.0])
failure = np.array([0.0,0.0,0.0])
for i in range(53):
    if y[i] == 0:
        failure = np.append(failure, X_pca[i])
    else:
        success = np.append(success, X_pca[i])
failure = failure.reshape((33,3))
success = success.reshape((22,3))
```

Stable: Orange, Unstable: Blue

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(failure[1:, 0], failure[1:, 1], failure[1:, 2], alpha=0.8, marker="x")
ax.scatter(success[1:, 0], success[1:, 1], success[1:, 2], alpha=0.5, marker="^")
ax.set_title("PCA")
```

## Apply K Nearest Neighbor

```python
score_train = 0
score_cross = 0
for i in range(1000):
    X_train, X_cross, y_train, y_cross = model_selection.train_test_split(X_pca, y, train_size=.8, random_state=i)
    neigh = neighbors.KNeighborsClassifier(n_neighbors=2)
    neigh.fit(X_train, y_train)
    score_train = score_train + neigh.score(X_train, y_train)
    score_cross = score_cross + neigh.score(X_cross, y_cross)
score_train = score_train / 1000
print('Train Accuracy: %.2f' % score_train)
score_cross = score_cross / 1000
print('Cross Accuracy: %.2f' % score_cross)
```
